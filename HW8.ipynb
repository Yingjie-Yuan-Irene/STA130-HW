## My Answer

### (a) The categories it solves with examples

The *Classification Decision Tree* is majorly used in solving *Classification Problems*, as known as dividing data into different categories. 

**Main Categories:**

***Binary Classification***
>
> Data is divided into two categories, such as "Yes/No" or "True/False".
>
> example: Determining whether an email is spam.

***Multi-class Classification***
>
> Data is divided into three or more categories.
>
> example: Categorizing customers into "high spender," "medium spender," or "low spender" based on purchasing behavior.

***Rule-based Segmentation Tasks***
>
> Creating explicit rules to segment data.
>
> example: Grouping students into "pass," "retest," or "fail" based on test scores

### (b) How It Performs Classification Predictions

A Classification Decision Tree uses feature values to progressively split data, ultimately categorizing it into a specific group.

**Workflow:**

Tree Construction:
> 
> 1. Divide data into subsets based on a feature (e.g., "Age" or "Income").
>
> 2. Repeat this process for each subset until a stopping condition is met (e.g., tree depth or purity).

Prediction Process:
>
> 1. Pass a new sample through the tree from the root to a leaf based on its feature values.
>
> 2. The category of the leaf is the prediction.

### (c) Comparison with Multiple Linear Regression (MLR)

**Model Complexity**: The complexity of the classification decision tree is reflected in the depth of the tree and the number of nodes, too deep a tree may lead to Overfitting. The complexity of multiple linear regression is reflected in the number of features included in the model and the problem of multicollinearity.

**Feature Interaction**: Classification decision trees naturally capture interactions between features because the decision path relies on the combination of multiple features. In contrast, multiple linear regression assumes that features act independently on the target variable, and if feature interactions need to be considered, Interaction Terms need to be explicitly included in the model.

- Decision Tree Structure: The model constructs a decision tree of depth 4 based on the feature variables NumPages, Thick and List Price.
- Splitting Nodes: At each node, the model selects the features and their thresholds that best separate the data for splitting. The choice of split is based on maximizing Information Gain or minimizing Gini Impurity.

- Decision Paths: When making predictions about new data, the model starts at the root node and moves down the branches of the tree based on the comparison of feature values to thresholds.

- Leaf Nodes: The model finally reaches the leaf node, which contains the predicted category, as well as the number of samples and the distribution of categories in the training set for that node.

- Prediction: The model outputs as a prediction the category that is dominant in the leaf nodes, i.e. whether the book is predicted to be Hard Cover or Paperback.

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>
  
> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>
> Compared to the simplicity of understanding how different **covariates** contribute towards the final **predicted values** of **multiple linear regression models** (by just reading off the equation to see how predictions work), the the complexity of how all the different **features** interact and combine to together to create the final **predictions** from **classification decision trees** can be staggering. But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions. Now we just need to be sure we're not **overfitting** our **classification decision trees** since they can be so **complex**. Fortunately, the "GridSearchCV" methodology mentioned in regards to finding the best `max_depth` setting for a tree is going to provide a general answer to the challenge of complexity and **overfitting** in **machine learning models** that is not too hard to understand (and which you might already have some guesses or a hunch about).
